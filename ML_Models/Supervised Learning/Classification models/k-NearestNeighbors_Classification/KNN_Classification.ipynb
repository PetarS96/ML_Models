# **KNN Model Theory**


## Theory
K-Nearest Neighbors (KNN) is a simple, instance-based learning algorithm used for classification and regression. For classification, KNN assigns a class label to a data point based on the majority class among its k-nearest neighbors in the feature space.

The main idea is to:
- Compute the distance between the input data point and all other points in the dataset.
- Select the k-nearest neighbors based on the smallest distances.
- Determine the majority class among these neighbors and assign it to the input point.

## Distance Metrics
The distance between two points is typically calculated using metrics like:
- **Euclidean Distance**: 
  $$ d(x, x') = \sqrt{\sum_{i=1}^{n} (x_i - x'_i)^2} $$
- **Manhattan Distance**: 
  $$ d(x, x') = \sum_{i=1}^{n} |x_i - x'_i| $$
- **Minkowski Distance** (generalization):
  $$ d(x, x') = \left(\sum_{i=1}^{n} |x_i - x'_i|^p \right)^{1/p} $$

## Classification Process

1. **Input**: 
   - Training data $(X, y)$ where $X$ are feature vectors and $y$ are corresponding class labels.
   - Test point $x_{\text{test}}$ to classify.
   - Hyperparameter $k$, the number of neighbors to consider.

2. **Compute Distances**: 
   Compute the distance between $x_{\text{test}}$ and all points in the training dataset.

3. **Sort and Select**: 
   Sort the distances and select the k smallest.

4. **Majority Voting**: 
   Determine the majority class among the k-nearest neighbors and assign it to $x_{\text{test}}$.

## Key Hyperparameters
- **k**: Determines the number of neighbors to consider. 
  - A small k may result in overfitting, while a large k can smooth the decision boundary but risk underfitting.
- **Distance metric**: Impacts the classification outcome based on the geometry of the feature space.

## Advantages
- Simple and intuitive.
- No explicit training phase (lazy learner).

## Disadvantages
- Computationally expensive for large datasets (distance computation for all points).
- Sensitive to irrelevant or scaled features (feature scaling is recommended).

## Example Update Rule
- No parameter updates are needed, as KNN is a non-parametric, lazy learning algorithm.



## **Model Evaluation for K-Nearest Neighbors (KNN) Classifier**

### 1. Accuracy Score

**Formula:**
$$
\text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}
$$

**Description:**
- **Accuracy** measures the proportion of correctly classified instances out of the total instances.
- Works well when the dataset is balanced.

**Interpretation:**
- A higher accuracy score indicates better overall performance.
- **Limitations:**  
  - Accuracy is not reliable for imbalanced datasets, where minority class performance may not be well-represented.

---

### 2. Precision

**Formula:**
$$
\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
$$

**Description:**
- **Precision** indicates the proportion of true positive predictions out of all positive predictions.
- Focuses on the reliability of positive predictions.

**Interpretation:**
- High precision indicates fewer false positives.
- **Use case:** Important in scenarios where false positives are costly, such as spam filtering.

---

### 3. Recall (Sensitivity)

**Formula:**
$$
\text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}
$$

**Description:**
- **Recall** measures the proportion of actual positive instances correctly identified.
- Indicates the ability of the model to capture all positive cases.

**Interpretation:**
- High recall indicates fewer false negatives.
- **Use case:** Critical in cases where missing positives is costly, such as medical diagnosis.

---

### 4. F1-Score

**Formula:**
$$
\text{F1} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

**Description:**
- **F1-Score** is the harmonic mean of precision and recall.
- Balances the trade-off between false positives and false negatives.

**Interpretation:**
- Higher F1-score indicates a good balance between precision and recall.
- **Use case:** Suitable for imbalanced datasets where both precision and recall matter.

---

### 5. Confusion Matrix

**Description:**
- A **confusion matrix** is a performance summary table showing:
  - True Positives (TP)
  - True Negatives (TN)
  - False Positives (FP)
  - False Negatives (FN)

**Interpretation:**
- Helps understand how well the classifier distinguishes between classes.
- Useful for identifying specific types of classification errors.

---

### 6. K-Fold Cross-Validation

**Description:**
- **K-Fold Cross-Validation** splits the dataset into `K` folds and evaluates the model `K` times, each time using a different fold as the test set and the remaining as the training set.

**Interpretation:**
- Provides a more robust estimate of model performance by reducing the bias of a single train-test split.
- Useful for validating the KNN model's stability across different subsets of the data.

---

### 7. ROC-AUC (Receiver Operating Characteristic - Area Under Curve)

**Description:**
- **ROC Curve:** Plots the true positive rate (TPR) against the false positive rate (FPR) at various thresholds.
- **AUC:** Represents the area under the ROC curve.

**Interpretation:**
- AUC ranges from 0.5 to 1.0:
  - **0.5:** Random performance.
  - **1.0:** Perfect classifier.
- Higher AUC values indicate better discrimination between classes.

---

### 8. Mean Distance (Optional KNN-specific metric)

**Description:**
- For KNN, the **mean distance** measures the average distance between a test sample and its `K` nearest neighbors.

**Interpretation:**
- Smaller mean distances indicate closer neighbors and potentially more accurate classifications.
- Useful for analyzing the neighborhood behavior of the model.
---
